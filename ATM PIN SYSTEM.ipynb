# ATM Cash Withdrawal System

```python
balance = 5000
pin = 1234

user_pin = int(input("Enter you PIN : "))

if user_pin == pin:
    amount = int(input("Enter amount:"))

    if amount <= balance:
        balance -= amount
        print("Withdrawal Successful")
        print("Remaining Balance :", balance)
    else:
        print("Insufficient Balance")
else:
    print("Incorrect PIN")
              
```

# New Feature Add

```python
balance = 5000
pin = 1234

attempts = 0
max_attempts = 3

while attempts < max_attempts:
    user_pin = int(input("Enter your PIN: "))

    if user_pin == pin:
        amount = int(input("Enter amount: "))

        if amount <= balance:
            balance -= amount
            print("Withdrawal Successful")
            print("Remaining Balance:", balance)
        else:
            print("Insufficient Balance")
        break   

    else:
        attempts += 1
        print("Incorrect PIN. Attempts left:", max_attempts - attempts)

        if attempts == max_attempts:
            print("Account Locked! Too many incorrect attempts.")
            print("Please visit your nearest bank branch.")
```
